```python
df['SepalLengthCm'].fillna(0, inplace=True)
df['SepalWidthCm'].fillna(0, inplace=True)
df['PetalLengthCm'].fillna(0, inplace=True)
df['PetalWidthCm'].fillna(0, inplace=True)

df.isnull().sum()
```

```python
df_encoded = pd.get_dummies(df, columns=['Species'])

df_encoded.columns

df_encoded['Species_Iris-setosa'] = df_encoded['Species_Iris-setosa'].astype(int)
df_encoded['Species_Iris-virginica'] = df_encoded['Species_Iris-virginica'].astype(int)
df_encoded['Species_Iris-versicolor'] = df_encoded['Species_Iris-versicolor'].astype(int)

df_encoded

# df_encoded.head()
```
