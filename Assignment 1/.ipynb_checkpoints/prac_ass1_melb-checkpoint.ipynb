```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('melb_data.csv')

df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
car_mean = df['Car'].mean()
buildingarea_mean = df['BuildingArea'].mean()
yearbuilt_mean = df['YearBuilt'].mean()
councilarea_mode = df['CouncilArea'].value_counts().idxmax()


df['Car'].fillna(car_mean, inplace=True)
df['BuildingArea'].fillna(buildingarea_mean, inplace=True)
df['YearBuilt'].fillna(yearbuilt_mean, inplace=True)
df['CouncilArea'].fillna(councilarea_mode, inplace=True)
 
df.isnull().sum()
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
df_encoded = pd.get_dummies(df, columns=['Regionname'])

df_encoded
```
