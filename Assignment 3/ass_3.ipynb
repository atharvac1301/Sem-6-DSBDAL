```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import datasets
import matplotlib.pyplot as plt

```

```python
iris = datasets.load_iris()   # iris has Summary Statistics

columns = ["sepal length", "sepal width", "petal length", "petal width"]

df = pd.DataFrame(iris.data, columns=columns)

df['target name'] = iris.target     # create a column of target names (one-hot encoded by default)

df
```

```python
df_setosa = df[df['target name']==0]
df_versicolor = df[df['target name']==1]
df_verginica = df[df['target name']==2]


# Changing one-hot codes to Original Names (Optional)
df_setosa.loc[:, 'target name'] = 'Iris-Setosa'
df_versicolor.loc[:, 'target name'] = 'Iris-Versicolor'
df_verginica.loc[:, 'target name'] = 'Iris-Verginica'

df_setosa
```

```python
# sl = sepal length, sw = sepal width, pl = petal length, pw = petal width

df_setosa 
df_versicolor 
df_verginica 


# Mean For Iris-Setosa:-

mean_setosa_sl, mean_setosa_sw = np.mean(df_setosa["sepal length"]), np.mean(df_setosa["sepal width"])
mean_setosa_pl, mean_setosa_pw = np.mean(df_setosa["petal length"]), np.mean(df_setosa["petal width"])

# Mean For Iris-Versicolor:-

mean_versicolor_sl, mean_versicolor_sw = np.mean(df_versicolor["sepal length"]), np.mean(df_versicolor["sepal width"])
mean_versicolor_pl, mean_versicolor_pw = np.mean(df_versicolor["petal length"]), np.mean(df_versicolor["petal width"])

# Mean For Iris-Verginica:-

mean_verginica_sl, mean_verginica_sw = np.mean(df_verginica["sepal length"]), np.mean(df_verginica["sepal width"])
mean_verginica_pl, mean_verginica_pw = np.mean(df_verginica["petal length"]), np.mean(df_verginica["petal width"])



# min_sl = np.min(df["sepal length"])
# min_sw = np.min(df["sepal width"])
# min_pl = np.min(df["petal length"])
# min_pw = np.min(df["petal width"])

# max_sl = np.max(df["sepal length"])
# max_sw = np.max(df["sepal width"])
# max_pl = np.max(df["petal length"])
# max_pw = np.max(df["petal width"])

# std_sl = np.std(df["sepal length"])
# std_sw = np.std(df["sepal width"])
# std_pl = np.std(df["petal length"])
# std_pw = np.std(df["petal width"])

```
