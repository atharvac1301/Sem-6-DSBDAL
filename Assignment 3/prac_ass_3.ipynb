```python
# Remove the last row since it has all NAN values
df.drop(457, inplace=True)

df
```

```python
frequent_college = df['College'].value_counts().idxmax()
mean_salary = df['Salary'].median()

df['College'].fillna(frequent_college, inplace=True)
df['Salary'].fillna(mean_salary, inplace=True)

df.isnull().sum()
```
