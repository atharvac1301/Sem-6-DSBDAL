```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Cabin'], axis=1)
df.head()
```

```python
sns.boxplot(data=df, x='Sex', y='Age', hue='Survived')
plt.show()
```

## Histogram for Males, Females and their Ages & Survival Status

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(data=df[(df['Sex']=='male')], x='Age', hue='Survived', multiple='stack', bins=50, ax=axes[0])
axes[0].set_title('Histogram for Males')

sns.histplot(data=df[(df['Sex']=='female')], x='Age', hue='Survived', multiple='stack', bins=50, ax=axes[1])
axes[1].set_title('Histogram for Females')

plt.show()
```

```python
survived_counts = df.groupby('Sex')['Survived'].sum()

print("Number of Women who survived:", survived_counts['female'])
print("Number of Men who survived:", survived_counts['male'])
```
