```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
sns.boxplot(x=df['Sex'], y=df['Age'], hue=df['Survived'])
```

# Correlation b/w Sex and Survied Status across Age Groups

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.histplot(data=df[df['Sex']=='male'], x="Age", hue="Survived", multiple='stack', bins=30, fill=True, ax=axes[0])
axes[0].set_title("Histogram for Males")

sns.histplot(data=df[df['Sex']=='female'], x="Age", hue="Survived", multiple='stack', bins=30, fill=True, ax=axes[1])
axes[1].set_title("Histogram for Females")

plt.show()
```
