# Outlier Detection using IQR (Inter Quartile Range)

```python
Q1 = np.percentile(df['reading score'], 25, interpolation='midpoint')    # 1st Quartile (25 %ile)
Q3 = np.percentile(df['reading score'], 75, interpolation='midpoint')    # 3rd Quartile (75 %ile)
IQR = Q3 - Q1    
```

```python
upper = Q3 + 1.5*IQR        # upper bound
lower = Q1 - 1.5*IQR        # lower bound
```

```python
# Remove Values (Outliers) that are below the Lower Bound and above the Upper Bound

df_filtered_IQR = df[(df['reading score'] <= upper) & (df['reading score'] >= lower)]

outliers_IQR = df[(df['reading score'] >= upper) | (df['reading score'] <= lower)]

df_filtered_IQR
```

```python
outliers_IQR
```

# Outlier Detection using Z-Score

```python
from scipy import stats

z_scores = stats.zscore(df['reading score'])   # Calculate Z-Score for each value in the dataset

z_scores
```

```python
z_threshold = 3

outliers_ZScore = df[abs(z_scores) >= z_threshold]

outliers_ZScore
```

```python
df_filtered_ZScore = df[abs(z_scores) < z_threshold]

df_filtered_ZScore
```

# Data Scaling

```python
# Min-Max Scaling

scaler = MinMaxScaler()

df_scaled = df    # copy the dataframe initially

df_scaled['reading score'] = scaler.fit_transform(df_scaled['reading score'].values.reshape(-1, 1))
df_scaled['writing score'] = scaler.fit_transform(df_scaled['writing score'].values.reshape(-1, 1))

df_scaled
```
