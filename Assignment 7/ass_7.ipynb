# TF-IDF

*TF-IDF reflects how important a word is to a document in a collection or corpus*

TF measures how a word appears in a document

**TF(t,d) = No of times t appears in document d / Total no. of terms in document d**

IDF measures how important a term is across the entire corpus. It penalizes common terms and gives more weight to rare terms.

**IDF(t,D) = log((Total no. of documents in the corpus N) / (Number of documents containing term t+1)) + 1** 

TF-IDF is calculated by - 

**TF-IDF(t, d, D) = TF(t,d) * IDF(t,D)**


```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = open('sample.txt', 'r')
documents = documents.read()
documents = sent_tokenize(documents)
documents
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

# Convert the TF-IDF matrix to a DataFrame for better readability
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=[f"Doc{i+1}" for i in range(len(documents))])

# Display the TF-IDF matrix
print("TF-IDF Matrix:")
print(df_tfidf)
```
