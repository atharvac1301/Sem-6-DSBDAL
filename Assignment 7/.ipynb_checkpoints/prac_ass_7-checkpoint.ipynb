```python

def tokenize(text):
    return word_tokenize(text)

tokens = tokenize(text)
print(tokens[0:150])

```

```python
tokens = tokenize(text)
pos_tag(tokens)
```

```python
def remove_stopwords(tokens):
    stopwords_list = stopwords.words('english')
    
    tokens = [token for token in tokens if token.lower() not in stopwords_list]
    return tokens

def stemming(tokens):
    ps = PorterStemmer()
    
    tokens = [ps.stem(token) for token in tokens]
    return tokens

def lemmatization(tokens):
    wnl = WordNetLemmatizer()
    
    tokens = [wnl.lemmatize(token) for token in tokens]
    return tokens

```

```python
def preprocess(tokens):
    tokens = remove_stopwords(tokens)
    tokens = stemming(tokens)
    tokens = lemmatization(tokens)
    
    return tokens

tokens = preprocess(tokens)
print(tokens[0:150])
    
    
```

## TF-IDF

#### *TF-IDF reflects how important a word is to a document in a collection or corpus*

![TF-IDF](https://miro.medium.com/v2/resize:fit:1400/format:webp/1*eBO93LqkFZW4DaZPRU2IUw.png)



```python
all_tokens = [None, None]
all_tokens[0] = tokenize(docs[0]) 
all_tokens[1] = tokenize(docs[1]) 

def tf(tokens):
    tf_dict = dict()
    
    for token in tokens:
        tf_dict[token] = tokens.count(token)
    
    for token in tf_dict:
        tf_dict[token] /= len(tokens)
    
    return tf_dict
            
tf(all_tokens[0])            
```
